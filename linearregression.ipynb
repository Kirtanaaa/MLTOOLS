# Simple Linear Regression 

y = ax1 + c

y = dependant variable

a = coefficient

x1 = independant variable (SLR has a single independant variable)

c = constant


**Linear Regression fits a linear model with coefficients = (w1, w2, w3,... wp) to minimize the residual sum of squares (RSS) between the observed targets in the dataset, and the targets predicted by the linear approximation**

Applications: No real world applications since it has only one independant variable and any real world application will have more than 1 independant variable.

## Importing Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing Data Set

```
ds=pd.read_csv("salary.csv")
x=ds.iloc[:,:-1].values
y=ds.iloc[:,-1].values
```

## Splitting into test and training set

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

## Training on Training Set using Simple Linear Regression

```
#try to find the best line between the given variables i.e apply linear regression to x_train and y_train
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

## Prediction on Test Set

```
y_pred=regressor.predict(x_test)
```

## Visualizing the Training Set

```
plt.scatter(x_train,y_train,color="lightgreen")
plt.plot(x_train,regressor.predict(x_train), color="red")
plt.title("CTC vs Years of Experience")
plt.xlabel("Experience in Years")
plt.ylabel("CTC")
plt.show()
```

## Visualizing the Test Set

```
plt.scatter(x_test,y_test,color="lightgreen")
plt.plot(x_train,regressor.predict(x_train), color="red")
plt.title("CTC vs Years of Experience")
plt.xlabel("Experience in Years")
plt.ylabel("CTC")
plt.show()
```
