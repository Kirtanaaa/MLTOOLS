```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.scatter(x,y)
plt.show()
```

```python
x1=[1,2,3,4]
y1=[84,63,16,57]
plt.scatter(x1,y1)
plt.plot(x1,y1)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]
plt.scatter(x,y,color="pink")
plt.scatter(x2,y2,color="red")
plt.plot(x,y,color="pink")
plt.plot(x2,y2,"r-.")
plt.show()
```

```python
a1=np.array([1,2,3,4,5,6])
b1=2*(a1**4)+100
plt.plot(a1,b1)
plt.show()
```

```python
a2=np.arange(0,10,0.1)
b2=2*(a2**4)+100
plt.plot(a2,b2)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="green", marker="*", linewidth=0.25)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
y1=x**3
plt.plot(x,y, color="green", marker="*", linewidth=0.25, label="x^4")
plt.plot(x,y1, color="red", marker="p", linewidth=0.25, label="x^3")
plt.xlabel("X - Axis",fontsize=15)
plt.ylabel("Y - Axis",fontsize=15)
plt.legend()
plt.title("MY GRAPHS PRACTICE")
plt.grid()
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
y1=x**3
plt.plot(x,y, color="green", marker="*", linewidth=0.25, label="x^4")
plt.plot(x,y1, color="red", marker="p", linewidth=0.25, label="x^3")
plt.xlabel("X - Axis",fontsize=15)
plt.ylabel("Y - Axis",fontsize=15)
plt.legend()
plt.title("MY GRAPHS PRACTICE")
plt.axis([0,15,0,20000])
plt.text(10,12500,"Aparna",fontsize=20)
plt.grid()
plt.show()
```

```python
profit=[10,20,30,25]
labels=["Amazon","Flipkart", "Jiomart", "Myntra"]
colors=["lightpink","lightyellow","lightblue","lightgreen"]
explode=[0.05,0.1,0,0]
plt.pie(profit,colors=colors,labels=labels,autopct="%.2f%%",explode=explode)
plt.show()
```
